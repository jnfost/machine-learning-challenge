```python
import pandas as pd
import numpy as np
import os
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv(os.path.join("Resources", "exoplanet_data.csv"))
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.info()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[["koi_fpflag_nt", "koi_fpflag_ss", "koi_fpflag_co", "koi_fpflag_ec", "koi_period", "koi_time0bk", "koi_impact", "koi_duration", "koi_depth", "koi_prad",
                       "koi_teq", "koi_insol", "koi_model_snr", "koi_tce_plnt_num", "koi_steff", "koi_slogg", "koi_srad", "ra", "dec", "koi_kepmag"]]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = df["koi_disposition"]
target_names = ["Candidate", "Confirmed", "False Positive"]

print(X.shape, y.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train).reshape(-1,1)
encoded_y_test = label_encoder.transform(y_test).reshape(-1,1)
```

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
y_train_categorical
```

# Train the Model



```python
from tensorflow.keras.models import Sequential
model3 = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 20

model3.add(Dense(units= 10, activation="relu", input_dim=number_inputs))
model3.add(Dense(units= 10, activation="relu"))
```

```python
number_classes = 3
model3.add(Dense(units=number_classes, activation="softmax"))
```

```python
model3.summary()
```

```python
model3.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model3.fit(
    X_train_scaled,
    y_train_categorical,
    epochs = 1000,
    shuffle = True,
    verbose = 1)
```

```python
model3_loss, model3_accuracy = model3.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Neural Network Loss: {model3_loss}, Accuracy: {model3_accuracy}")
```

```python
predictions = model3.predict_classes(X_test_scaled)
prediction_labels = label_encoder.inverse_transform(predictions)
prediction_labels
```

```python
print(f"Predicted Classes: {prediction_labels[:100:10]}")
print(f"Actual Classes: {list(y_test[:100:10])}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions, target_names=target_names))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {"C": [1,5,10],
             "gamma": [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model3, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
